```python
import pandas as pd
```

```python
df = pd.read_csv('admission_summary.csv')
```

```python
df.isna().sum()
```

```python
df['patient_name'] = (
    df['patient_name']
    .str.strip()
    .str.title()
)
```

```python
df['gender'] = df['gender'].str.strip().str.lower()
```

```python
df['gender'] = df['gender'].replace({
    'male': 'Male',
    'female': 'Female'
})
```

```python
df['age'] = pd.to_numeric(df['age'], errors='coerce')
```

```python
df['admission_date'] = pd.to_datetime(df['admission_date'])
df['discharge_date'] = pd.to_datetime(df['discharge_date'])
```

```python
df[['admission_date', 'discharge_date']].isna().sum()
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop_duplicates(
    subset=['patient_name', 'admission_date', 'diagnosis_name']
)
```

```python
df['calculated_los'] = (
    df['discharge_date'] - df['admission_date']
).dt.days
```

```python
df.to_csv('admission_summary_cleaned.csv', index=False)
```
